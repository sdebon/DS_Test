# Matplotlib
Matplotlib is a plotting library for Python. 
More examples: https://matplotlib.org/gallery/index.html

```python
# Import the libs
import numpy as np
import matplotlib.pyplot as plt
```

Script for generaing the plot for the equation y = 2x + 1

```python
X = np.array([2,3,4,5,6,7])
Y = np.array([5,6,5,9,7,10])

sum_x = 0
for elem in X:
    sum_x += elem
    
sum_x2 = 0
for elem in X:
    print (elem**2)
    sum_x2 += (elem**2)
    
sum_xy = 0
for elem in range(len(X)):
        sum_xy += X[elem]*Y[elem]

sum_y = 0
for elem in Y:
    sum_y += elem

x = np.arange(-3,10,1)
y = 5 + 1.46105*x
    
print(sum_x, sum_y, 'x^2:', sum_x2, 'xy:', sum_xy)
        

plt.xlabel("x axis") # add x axis caption
plt.ylabel("y axis") # add y axis caption
plt.plot(X,Y, 'bo') # draw the plot
plt.plot(y)
plt.show() # show the plot
```

```python
# Another example for y = x^3
x = np.arange(-1,1.05,0.001) # x points from -1 to 1.05 with step 0.05
y = x ** 3 # y points
plt.title("y = x^3") # add title
plt.xlabel("x axis") # add x axis caption
plt.ylabel("y axis") # add y axis caption

# add vertical / horizental lines
plt.vlines(0, -0.5, 0.5, colors = "g", linestyles = "dashed")
plt.hlines(0, -1, 1, colors = "g", linestyles = "dashed")

plt.plot(x,y) 
plt.show()
```

Plot different things in the same figure. 

```python
# domain of x
x = np.arange(0, 3 * np.pi, 0.1)

# setup the subplot grid that has two rows and one column
# set the first subplot as active
plt.subplot(2, 1, 1)

# make the first plot
plt.plot(x, np.sin(x))
plt.title('sin(x)')

# set the second subplot as active
plt.subplot(2, 1, 2)

# make the second plot
plt.plot(x, np.cos(3*x))
plt.title('cos(x)')


# show the figure.
plt.show()
```
